# Empty dictionary

```python
df = {"Title" : [], "Price": [], "Area":[], "Area-Type": []}
```

```python
df
```

# Price

```python
for listing in soup.find_all(class_ = "mb-srp__card__estimate"):
    price = listing.select(".mb-srp__card__price--amount")[0].text.strip()
    df["Price"].append(price)
```

```python
len(df["Title"])
```

# Area & AreaType

```python
listings = soup.find_all('div', class_ = "mb-srp__card__summary__list")
```

```python
for listing in listings:
    labels = listing.find_all('div',class_ = "mb-srp__card__summary--label")
    values = listing.find_all('div',class_ = "mb-srp__card__summary--value")
    for label,value in zip(labels,values):
        label_text = label.text.strip()
        value_text = value.text.strip()
        if "Carpet Area" in label_text:
            df["Area-Type"].append("Carpet Area")
            df["Area"].append(value_text)
        elif "Super Area" in label_text:
            df["Area-Type"].append("Super Area")
            df["Area"].append(value_text)
```

# Title

```python
soup.find_all("h2",class_ = "mb-srp__card--title")[0].text.strip()
```

```python
for title in soup.find_all("h2",class_ = "mb-srp__card--title"):
    df["Title"].append(title.text.strip())
```

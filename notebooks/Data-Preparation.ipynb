```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("../data/raw/real_estate.csv")
```

```python
df.drop_duplicates(keep = "first", inplace = True)
```

```python
df["Area"] = [i[0] for i in df["Area"].str.split()]
df["Area"] = df["Area"].str.replace(",","")
df["Area"] = df["Area"].astype(float)
df["Area"]
```

```python
df.loc[df['Area-Type'] == 'Carpet Area', 'Area'] *= 1.5
df.drop("Area-Type",axis=1,inplace = True)
```

```python
df
```

```python
df["BHK"] = [i[0] for i in df["Title"].str.split()]
```

```python
df["BHK"].unique()
```

```python
df.loc[df["BHK"].isin(['Builder','Apartment','Studio','30490', '3500', '3484', '1100', '4400', '4356', '4000','5200', '4360', '2000', '1750', '2485', '2178', '2200', '1350']),"BHK"]=0
```

```python
df[df["BHK"] == 0]
```

```python
df["BHK"] = df["BHK"].replace("4+","6")
df["BHK"] = df["BHK"].astype(int)
df["BHK"].unique()
```

```python
df["Location"] = [i[-2].replace(",","") for i in df["Title"].str.split()]
```

```python
df.loc[df["Location"] == "in","Location"] = np.nan
df["Location"] = df["Location"].replace("Hill","Mary Hill")
df["Location"] = df["Location"].replace("Hills","Kadri")
df["Location"] = df["Location"].replace("Hills","Kadri")
df["Location"] = df["Location"].replace("Estate","Yeyyadi")
df["Location"] = df["Location"].replace("Chowk","Kottara")
df["Location"] = df["Location"].replace("Street","Car Street")
df["Location"] = df["Location"].replace("Shivabaugh","Shivabagh")
df["Location"] = df["Location"].replace("Kambla","Kadri")
df["Location"] = df["Location"].replace("bagh","Shivabagh")
df.loc[df["Location"] == "Junction","Location"] = "Kuttar"
df.loc[df["Location"] == "Lane","Location"] = "Vas Lane"
df.loc[df["Location"] == "hostel","Location"] = "Kaprigudda"
df.loc[df["Location"] == "Bail","Location"] = "Hoige Bail"
df.loc[df["Location"] == "Derelakatai","Location"] = "Derlakatte"
df.loc[df["Location"] == "Kodialbail","Location"] = "Kodailbail"
#df["Location"] = df["Location"].replace(",","",regex = True)
```

```python
df.loc[df["Title"].str.contains("Flat") | df["Title"].str.contains("Apartment"),"Type"] = "Flat"
df.loc[df["Title"].str.contains("Villa"),"Type"] = "Villa"
df.loc[df["Title"].str.contains("Plot"),"Type"] = "Plot"
```

```python
df = df[df["Type"].notnull()]
```

```python
df = df.drop("Title",axis =1)
```

```python
df["Price"] = df["Price"].replace("₹","",regex=True)
```

```python
df["Price"] = df["Price"].fillna("0")
```

```python
df["Price"]
```

```python
def convert(value):
    if "Lac" in value:
        return float(value.split()[0]) * 1e5
    elif "Cr" in value:
        return float(value.split()[0]) * 1e7
    else:
        return np.nan
```

```python
df["Price"] = df["Price"].replace("Lakh","Lac", regex = True)
df["Price"] = df["Price"].replace("Lakhs","Lac",regex = True)
df["Price"] = df["Price"].replace("Crore","Cr",regex = True)
df["Price"] = df["Price"].replace("Lacs","Lac",regex = True)
df["Price"]
```

```python
df["Price_num"] = df["Price"].apply(convert)
```

```python
df.isna().sum()
```

```python
df["Location"] = df["Location"].fillna("Others")
```

```python
df["Location"].unique()
```

```python
df
```

```python
df[df["Area"]/df["BHK"] <300]
df = df[df["Area"]/df["BHK"] > 300]
```

```python
df["Price/Sqft"] = round(df["Price_num"]/df["Area"])
```

```python
df
```

```python
df.to_csv("../data/transformed/Transformed_data.csv",index = False)
```
